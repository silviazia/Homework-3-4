Read the Whale Portfolio daily returns and clean the data.

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data.

```python
# Reading algorithmic returns
df_algorithmic_returns = pd.read_csv("Data/algo_returns.csv",index_col='Date',parse_dates=True,infer_datetime_format=True)
```

```python
df_algorithmic_returns.head(5)
```

```python
# Count nulls
df_algorithmic_returns.isnull().sum()
```

## S&P TSX 60 Returns

Read the S&P TSX 60 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P TSX 60 Closing Prices
df_sp_tsx_history = pd.read_csv("Data/sp_tsx_history.csv",index_col='Date',parse_dates=True,infer_datetime_format=True)
```

```python
# Check Data Types
df_sp_tsx_history["Close"].dtype
```

```python
# Fix Data Types
df_sp_tsx_history["Close"] = df_sp_tsx_history["Close"].str.replace("$", "")
df_sp_tsx_history["Close"] = df_sp_tsx_history["Close"].str.replace(",", "")
df_sp_tsx_history["Close"] = df_sp_tsx_history["Close"].astype("float")
df_sp_tsx_history["Close"].head(10)
```

```python
# Calculate Daily Returns
df_sp_tsx_returns = df_sp_tsx_history.pct_change()
df_sp_tsx_returns.head()
```

```python
# Drop nulls
df_sp_tsx_returns.dropna(inplace=True)
```

```python
df_sp_tsx_returns.head()
```

```python
# Rename `Close` Column to be specific to this portfolio.
df_sp_tsx_returns = df_sp_tsx_returns.rename(columns={"Close": "SP_TSX60 Returns"})
```

```python
df_sp_tsx_returns.head()
```

## Combine Whale, Algorithmic, and S&P TSX 60 Returns

```python
# Join Whale Returns, Algorithmic Returns, and the S&P TSX 60 Returns into a single DataFrame with columns for each portfolio's returns.
all_portfolio_return = pd.concat([df_whale_returns, df_algorithmic_returns, df_sp_tsx_returns], axis='columns', join='inner')
all_portfolio_return.head()
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Calculate and plot the rolling standard deviation for the S&P TSX 60 using a 21-day window.
2. Calculate the correlation between each stock to determine which portfolios may mimick the S&P TSX 60.

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
all_portfolio_rolling_std = all_portfolio_return.rolling(21).std()

# Plot the rolling standard deviation
all_portfolio_rolling_std.plot(figsize=(20, 10), title="21-day Rolling Std")
```

### Calculate and plot the correlation

```python
# Calculate the correlation
corr_df = all_portfolio_return.corr()
# Display de correlation matrix
corr_df.style.background_gradient(cmap="viridis")
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 60 TSX

```python
# Calculate covariance of a single portfolio
covariance = all_portfolio_return["Algo 1"].rolling(60).cov(all_portfolio_return["SP_TSX60 Returns"])

# Calculate variance of S&P TSX
variance = all_portfolio_return["SP_TSX60 Returns"].rolling(60).var()

# Computing beta
beta = covariance / variance

# Plot beta trend
beta.plot(figsize=(20, 10), title="Algo 1 beta")
```

## Rolling Statistics Challenge: Exponentially Weighted Average 

An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the [`ewm`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html) with a 21-day half-life.

```python
# Use `ewm` to calculate the rolling window
all_portfolio_return.ewm(halflife=21).std().head()
```

```python
all_portfolio_return.plot(figsize=(20, 10), title="Exponentially Weighted Average")
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
sharpe_ratios = (all_portfolio_return.mean()*252) / (all_portfolio_return.std() * np.sqrt(252))
sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot.bar(figsize=(20, 10), title="Sharpe Ratios")
```

## Choose 3-5 custom stocks with at last 1 year's worth of historic prices and create a DataFrame of the closing prices and dates for each stock.

For this demo solution, we fetch data from three companies listes in the S&P TSX 60 index.

* `SHOP` - [Shopify Inc](https://en.wikipedia.org/wiki/Shopify)

* `OTEX` - [Open Text Corporation](https://en.wikipedia.org/wiki/OpenText)

* `L` - [Loblaw Companies Limited](https://en.wikipedia.org/wiki/Loblaw_Companies)

```python
# Reading data from shop stock
shop_historical_path = Path('Data/shop_historical.csv')
shop_historical_df = pd.read_csv(shop_historical_path, index_col='Date',parse_dates=True,infer_datetime_format=True, usecols=["Date", "Close"])
shop_historical_df.head()
```

```python
# Reading data from otex stock
otex_historical_path = Path('Data/otex_historical.csv')
otex_historical_df = pd.read_csv(otex_historical_path, index_col='Date',parse_dates=True,infer_datetime_format=True, usecols=["Date", "Close"])
otex_historical_df.head()
```

```python
# Reading data from L stock
l_historical_path = Path('Data/l_historical.csv')
l_historical_df = pd.read_csv(l_historical_path, index_col='Date',parse_dates=True,infer_datetime_format=True, usecols=["Date", "Close"])
l_historical_df.head()
```

```python
# Combine all stocks in a single DataFrame
my_portfolio = pd.concat([shop_historical_df, otex_historical_df, l_historical_df], axis='columns', join='inner')
my_portfolio.columns = ["Shop", "Otex", "L"]
my_portfolio_return = my_portfolio.sort_index().pct_change()
my_portfolio_return.head()
```

```python
# Reset Date index
my_portfolio_return.dropna(inplace=True)
my_portfolio_return.head()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Set weights
weights = [1/3, 1/3, 1/3]
weighted_portfolio = my_portfolio_return.dot(weights)
weighted_portfolio.sum()
```

```python
# Cumulative Returns of My Portfolio
my_cumulative_returns = (weighted_portfolio + 1).cumprod()
my_cumulative_returns.sum()
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Join your returns DataFrame to the original returns DataFrame
my_portfolio_return = pd.concat([weighted_portfolio, df_whale_returns, df_algorithmic_returns, df_sp_tsx_returns], axis='columns', join='inner')
my_portfolio_return.rename(columns={0:"My Portfolio"}, inplace=True)
my_portfolio_return.head()
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)
my_portfolio_return.dropna(inplace=True)
my_portfolio_return.head()
```

## Re-run the risk analysis with your portfolio to see how it compares to the others
